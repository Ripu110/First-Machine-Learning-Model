#LOAD DATASET

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('delaney_solubility_with_descriptors.csv')
```

```python
df
```

###Data Preparation

```python
#Data Preparation as X and Y.
```

```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis = 1)
x
```

###Data Splitting

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 100)
```

```python
x_train
```

```python
x_test
```

###***Data Modelling***

###**Linear Regression**

#Training the model for linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

#Applying the model for linearRegression

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)


```

```python
print(y_lr_train_pred, y_lr_test_pred)
```

#Evaluate Model Performance

```python
y_train
```

```python
y_lr_train_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
lr_train_mse
```

```python
lr_test_mse
```

```python
lr_train_r2
```

```python
lr_test_r2
```

```python
print('LR MSE (Train):', lr_train_mse)
print('LR MSE (Test):', lr_test_mse)
print('LR R2 (Train):', lr_train_r2)
print('LR R2 (Test):', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
```

```python
lr_results
```

```python
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

###**Random Forest**

#Training the Model

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth = 2, random_state = 100)
rf.fit(x_train, y_train)
```

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

```python
y_rf_train_pred
```

```python
y_rf_test_pred
```

###**Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
rf_results
```

###**Model Comparison**

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
df_models
```

```python
# @title Test MSE vs Test R2

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Test R2'].value_counts()
    for x_label, grp in df_models.groupby('Test MSE')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Test MSE')
_ = plt.ylabel('Test R2')
```

```python
# @title Method

from matplotlib import pyplot as plt
import seaborn as sns
df_models.groupby('Method').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```
